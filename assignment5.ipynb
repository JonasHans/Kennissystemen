```python
from nltk.corpus import wordnet as wn

nouns = ['tree.n.01','forest.n.01','body.n.01']

for noun in nouns:
    print "Showing information on ",noun
    print "Part Meronyms : ",str(wn.synset(noun).part_meronyms())
    print "Member Meronyms : ",str(wn.synset(noun).member_meronyms())
    print "Substance Meronyms : ",str(wn.synset(noun).member_meronyms())
    print "Part Holonyms : ",str(wn.synset(noun).part_holonyms())
    print "Member Holonyms : ",str(wn.synset(noun).member_holonyms())
    print "Substance Holonyms : ",str(wn.synset(noun).member_holonyms()),"\n"
```

### Q13

```python
from nltk.corpus import wordnet as wn
from __future__ import division

nouns = wn.all_synsets('n')
total = 0
hyponyms = 0

for noun in nouns:
    if not noun.hyponyms():
        hyponyms += 1
    total += 1
        
print "Total number of nouns",total
print "Number of nouns with hyponyms ",hyponyms
print "Percentage of nouns with hyponyms ",round((hyponyms/total)*100,2),"%"
```

### Q14

```python
from nltk.corpus import wordnet as wn

def supergloss(s):
    print "Definition of synset : ",wn.synset(s).definition(),"\n"
    
    print "Definitions of all hypernyms :"
    for hypernym in wn.synset(s).hypernyms():
        print hypernym.definition()
        
    print "\nDefintions of all Hyponyms : "
    for hyponym in wn.synset(s).hyponyms():
        print hyponym.definition()
        
supergloss('tree.n.01')
```

### Q26

```python
from nltk.corpus import wordnet as wn
from __future__ import division

nouns = wn.all_synsets('n')
nounsWithHyponyms = 0
hyponyms = 0

for noun in nouns:
    if noun.hyponyms():
        hyponyms += len(noun.hyponyms())
        nounsWithHyponyms += 1
    
print "Number of nouns with hyponyms are ", nounsWithHyponyms
print "Average number of hyponyms per noun is ", round(hyponyms/nounsWithHyponyms,2)
```

### Q27

```python
from nltk.corpus import wordnet as wn
from __future__ import division

def averagePolysems(typeName, typeKey):
    nouns = wn.all_synsets(typeKey)
    numberOfNounSenses = 0
    numberOfNouns = 0

    for noun in nouns:
        numberOfNounSenses += len(wn.synsets(noun.name().split('.')[0],typeKey))
        numberOfNouns += 1

    print "Average polysemy of ", typeName," : ",round(numberOfNounSenses/numberOfNouns, 2)
    
averagePolysems("nouns", "n")
averagePolysems("verbs", "v")
averagePolysems("adjectives", "a")
averagePolysems("adverbs", "r")
```

## My Spotlight answers

```python
import requests 
import json
from bs4 import BeautifulSoup

def inspectURL(url, website):
    # Get url and text     
    bbc= requests.get(url)
    soup= BeautifulSoup(bbc.text) 

    if website == "wiki":
        article=soup.find("div", { "id" : "mw-content-text" })
    else:
        article=soup.find('div', {'class':'story-body__inner'})
        
    article_text= article.text
    article_text[:2000]
    confidence = raw_input('What should the confidence value be? ')
    support = raw_input('What should the support value be? ')
    queryDbpedia(article_text, confidence, support)

def queryDbpedia(article_text, confidence, support):
    # Set header and body for spotlight API request      
    spotlight = "http://spotlight.dbpedia.org/rest/annotate"
    headers = { "Accept" : "application/json", "Content-type" : "application/x-www-form-urlencoded;charset=utf-8" }
    form= { "text" : article_text, "confidence" : confidence, "support": support}
    
    #Doe de POST request
    ef= requests.post(spotlight,headers=headers,data=form)
    results = {}
    try:
        words = ef.json()["Resources"]
        print "Significant words :\n"
        for word in words:
            results[word["@surfaceForm"]]= word["@URI"] 
        for key in sorted(results):
            print key+" -> "+results[key]
        print "\nNumber of entities found is "+str(len(sorted(results)))+"\n"
    except:
        print "Spotlight API not available, trying to connect again...\n"
        queryDbpedia(article_text, confidence, support)
    
    return results
        
print "Article 1 : https://en.wikipedia.org/wiki/University_of_Amsterdam"
inspectURL("https://en.wikipedia.org/wiki/University_of_Amsterdam", "wiki")
print "Article 2 : https://en.wikipedia.org/wiki/Mount_Olympus"
inspectURL("https://en.wikipedia.org/wiki/Mount_Olympus", "wiki")
```
