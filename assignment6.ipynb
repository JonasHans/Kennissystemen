## Structure decisions


 In the xml structure a children structure was used because with this generalities could be displayed well. We chose to divide all words in to their functions, in this case noun. In our xml structure etymology is defined in various children relations. There is a root of the word, which has the word and the language it is in. But because this word is derived from another word from another language, there is another root a level deeper which has the word and language from which the first word was derived. The level keeps going deeper until we get the lowest form of which the word was derived. Furthermore a word has meanings, every meaning can have several definitions and every definition can have several subdefinitions. This cleary displays the relation between all the different levels of meaning of a word.
